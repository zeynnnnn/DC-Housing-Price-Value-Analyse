```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("DC_Properties.csv")
```

```python
data= data.dropna(subset= ["SALEDATE","PRICE"])
data = data.drop(columns="Unnamed: 0")
data = data.drop_duplicates()
```

```python
import datetime
from datetime import datetime
days=["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]
weekdaySeries=pd.DataFrame()
weekdaySeries["Weekdays"]=data["SALEDATE"]
weekdaySeries["PRICE"]=data["PRICE"]
for i in data.index:
    weekdaySeries["Weekdays"][i] = days[datetime.strptime(data["SALEDATE"][i],'%Y-%m-%d %H:%M:%S').weekday()]
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import colors
from matplotlib.ticker import PercentFormatter
```

```python
plt.figure(figsize=(8,8))
target_count = weekdaySeries.groupby(weekdaySeries["Weekdays"])["Weekdays"].count()
target_count.plot(kind='bar', title='Sale Number vs Days of Week',grid=True, color='#607c8e')
plt.xlabel('Days of Week')
plt.ylabel('Number of House Sale')
plt.tight_layout()
plt.savefig("saleNum.svg")
plt.savefig("saleNum.png")
plt.show()
```

```python
plt.figure(figsize=(8,8))
target_count = weekdaySeries.groupby(weekdaySeries["Weekdays"])["PRICE"].mean()
target_count.plot(kind='bar', title='Mean Price of Sold Houses vs Days of Week',grid=True, color='#607c8e')
plt.xlabel('Days of Week')
plt.ylabel('Mean Price of Sold Houses($)')
plt.tight_layout()
plt.savefig("price_weekday.svg")
plt.savefig("price_weekday.png")
plt.show()
```
