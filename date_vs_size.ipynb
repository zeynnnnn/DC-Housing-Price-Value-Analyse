```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import colors
from matplotlib.ticker import PercentFormatter
```

```python
plt.figure(figsize=(8,8))
target_count = data.groupby(pd.cut(data["LIVING_GBA"], np.arange(0,8553, 100)))[["AYB","LIVING_GBA"]].mean()
target_count.plot(x=  "LIVING_GBA" ,y="AYB",kind='scatter', title='Gross Building Area (ft²) VS Earliest Time Building BuilT',grid=True, color='#607c8e')
plt.xlabel('Gross Building Area (ft²)')
plt.ylabel('Earliest Time Building Built')
plt.show()
```

```python
plt.figure(figsize=(8,8))
data[["AYB","LIVING_GBA"]].plot(x=  "AYB" ,y="LIVING_GBA",kind='scatter', title='Earliest Time Building Built vs Gross Building Area (ft²)',grid=True, color='#607c8e')
plt.xlabel('Earliest Time Building Built')
plt.ylabel('Gross Building Area (ft²)')
plt.tight_layout()
plt.savefig("area_scatter.png")
plt.show()
```

```python
pd.cut(data["AYB"], np.arange(1780,2018, 3))
```

```python
plt.figure(figsize=(18,8))
target_count = data.groupby(data["AYB"])["LIVING_GBA"].mean()
target_count.plot(kind='bar', title='Earliest Time Building Built vs Gross Building Area',grid=True, color='#607c8e')
plt.xlabel('Earliest Time Building Built')
plt.ylabel('Gross Building Area (ft²)')
plt.tight_layout()
plt.savefig("area_bar.png")
plt.show()
```
